```python
# remove nan values 
nan_count = np.sum(df.isnull(), axis = 0)
nan_count

```

```python
#change host acceptance and response rate to 0 instead od nan
df['host_response_rate'].fillna(0.0, inplace=True)
df['host_acceptance_rate'].fillna(0.0, inplace=True)
```

```python
#update feature list
non_feature_list = df.select_dtypes(include=['object']).columns
df = df.drop(columns=non_feature_list)
```

```python
#fit model and training data
y = df['instant_bookable']
X = df.drop(columns = 'instant_bookable', axis =1)
```

```python
# create training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=1234)
```

```python
#inspect training and test dat
print(X_train.shape)
print(X_test.shape)
```

```python
X_train.head()
```

```python
#function  to evalaute model 
def train_test_DT(X_train, X_test, y_train, y_test, depth, leaf=1, crit='entropy'):
    # Initialize the Decision Tree Classifier 
    model = DecisionTreeClassifier(max_depth=depth, min_samples_leaf=leaf, criterion=crit)
    
    # Fit the model on training data
    model.fit(X_train, y_train)
    
    # Predict the class labels for the test set
    class_label_predictions = model.predict(X_test)
    
    # Calculateaccuracy score
    acc_score = accuracy_score(y_test, class_label_predictions)
    
    return acc_score
    
```

```python
#perform decision tree model selection
max_depth_range = [2,8]
acc_scores = []
for md in max_depth_range:
    score = train_test_DT(X_train, X_test, y_train, y_test, md)
    acc_scores.append(float(score))
```

```python
#perform grid search to improve model by finding best hyperparameter values
param_grid = {'max_depth':[2**n for n in range(2,5)] , 'min_samples_leaf':[25*2**n for n in range(0,3)]}
param_grid
```

```python
#perform grid_search
model_grid = DecisionTreeClassifier()
grid = GridSearchCV(model_grid, param_grid, cv=5)
grid_search=grid.fit(X_train, y_train)
```

```python
#find the best hyperparameter
print('best hyperparameters: {0}'.format(grid_search.best_params_))
#print accurarcy score
print('Accuracy score: {0}'.format(grid_search.best_score_))
```
